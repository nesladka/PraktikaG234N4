```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('titanic.csv')
df
```

```python

df.index
```

```python
df.values
```

```python
passenger_counts = df['PClass'].value_counts().sort_index()

passenger_counts.plot(kind='bar', color='lavender')
plt.xlabel('Класс пассажиров')
plt.ylabel('Количество пассажиров')
plt.title('Зависимость количества пассажиров от класса')
plt.xticks(rotation=0)
plt.show()
```

```python
values = ('Женский', 'Мужской')
died_people = df[df['Survived'] == 0].groupby('Sex')['Survived'].count()
```

```python
plt.pie(died_people, labels = values, autopct='%1.1f%%', shadow=True, wedgeprops={'lw':1,'edgecolor':"k"}, colors=['lavender', 'lightskyblue', 'lightgreen'])
plt.title('Количество невыживших в процентном соотношении в зависимости от пола')
plt.show()
```
